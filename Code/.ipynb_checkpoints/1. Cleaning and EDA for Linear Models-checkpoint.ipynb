### Because there are so many rows we want to drop, lets make a function for it

````python
def drop_rows(column_name, to_drop):
    train.drop( train[train[column_name] > to_drop].index, inplace = True)
````

````python
drop_rows('Garage Yr Blt', 2014)
drop_rows('Fireplaces', 3)
drop_rows('Full Bath',3)
drop_rows('Bedroom AbvGr', 7)
drop_rows('Full Bath', 3)
drop_rows('Bsmt Half Bath', 1)
drop_rows('Bsmt Full Bath',2)
drop_rows('Gr Liv Area', 5000)
drop_rows('1st Flr SF', 3500)
drop_rows('Total Bsmt SF', 5000)
drop_rows('BsmtFin SF 1',3000)
drop_rows('Lot Area',70_000)
drop_rows('Lot Frontage', 150)
drop_rows('Kitchen AbvGr', 2)
````

# Getting Ready for some Modeling <span id='test1'>

````python
X_1 = train_copy[corr_features]
y_1 = train_copy['SalePrice']
````

````python
X_1_train, X_1_test, y_1_train, y_1_test = train_test_split(X_1, y_1, test_size = .33, random_state=777)
# We have an okay size data set, so let's try splitting it a bit more than 25%
# also setting a random_state so that i will have reproducible results, just for now. 
print(X_1_train.shape)
print(X_1_test.shape)
````

````python
# Lets get our first score
lr = LinearRegression() 
cross_val_score(lr, X_1, y_1)
````

````python
## Lets check based on our train model
lr.fit(X_1_train, y_1_train)
train_1_r2 = lr.score(X_1_train, y_1_train)
test_1_r2 = lr.score(X_1_test, y_1_test)

print(f'Train R2 Score: {train_1_r2}')
print(f'Test R2 Score : {test_1_r2}')

# very similar, good to know
````

````python
### Code inspiration with some modifications taken from Noah's follow along breakfast hour
## Lets also get the rmse
y_train_baseline = [y_1_train.mean()]*len(y_1_train)
y_test_baseline  = [y_1_train.mean()]*len(y_1_test)
# here are just make the variables for the mean, assuming all else is equal in terms of predicting the train and test,
# based on the train mean, for however long the y_train and y_test lengths are. 

# this will be for the train 
rmse_1 = metrics.mean_squared_error(y_1_train, y_train_baseline, squared = False)
print(f'Train RMSE Baseline: {rmse_1}')

# this will be for the test
rmse_1 = metrics.mean_squared_error(y_1_test, y_test_baseline, squared = False)
print(f'Test RMSE Baseline: {rmse_1}')
````

````python
# lets do both mse and rmse here
# first lets get our predictions
y_1_hat = lr.predict(X_1_train)
rmse_1 = metrics.mean_squared_error(y_1_train, y_1_hat, squared= False)
print(f'Train RMSE : {rmse_1}')

y_1_hat_test = lr.predict(X_1_test)
rmse_1 = metrics.mean_squared_error(y_1_test, y_1_hat_test, squared= False)
print(f'Test RMSE : {rmse_1}')
````

**Not bad, went down by almost 6 times on the MSE and a little over 2 on the RSME**

## Lets transform our new Numeric Features into Polynomial Features <span id='polyf'>

````python
# Lets try Polynomial features
num_features = train._get_numeric_data().columns
X_2 = train[num_features].drop('SalePrice', axis = 1)
poly = PolynomialFeatures(degree = 2, include_bias= False)

# Lets fit and transform
X_poly = poly.fit_transform(X_2)

X_2_poly = pd.DataFrame(data = X_poly, columns = poly.get_feature_names(num_features), index= X_2.index)
X_2_poly.head()
````

````python
# Do the same for the test set
num_features = train._get_numeric_data().drop('SalePrice', axis = 1).columns
kagg_X_2 = test[num_features]

kagg_poly = PolynomialFeatures(degree = 2, include_bias=False)

# Lets fit and transform
kagg_X_poly = kagg_poly.fit_transform(kagg_X_2)

kagg_X_2_poly = pd.DataFrame(data = kagg_X_poly, columns = poly.get_feature_names(num_features), index = kagg_X_2.index)
kagg_X_2_poly.head()
````

````python
lr_2 = LinearRegression()
y = train['SalePrice']

print(f'Before Poly: {cross_val_score(lr_2, X_2, y)}')
print(f'After Poly: {cross_val_score(lr_2, X_2_poly, y)}')
````

### Lets do a test run with X_2_poly and y <span id ='test2'>

````python
X_2_train, X_2_test, y_2_train, y_2_test = train_test_split(X_2_poly, y, random_state=770)
````

````python
# Getting some scores for our X_2 
y_train_baseline = [y_2_train.mean()]*len(y_2_train)
y_test_baseline  = [y_2_train.mean()]*len(y_2_test)

# this will be for the train 
rmse_2 = metrics.mean_squared_error(y_2_train, y_train_baseline, squared = False)
print(f'Train RMSE Baseline: {rmse_2}')

# this will be for the test
rmse_2 = metrics.mean_squared_error(y_2_test, y_test_baseline, squared = False)
print(f'Test RMSE Baseline: {rmse_2}')
````

````python
# FIT THE MODEL WITH THE TRAIN
lr_2.fit(X_2_train, y_2_train)
````

````python
# Rmse time
# first lets get our predictions
print(f'R2 Train Score: {lr_2.score(X_2_train, y_2_train)}')
print(f'R2 Test Score: {lr_2.score(X_2_test, y_2_test)}')

y_2_hat = lr_2.predict(X_2_train)
rmse_2 = metrics.mean_squared_error(y_2_train, y_2_hat, squared= False)
print(f'\nTrain RMSE : {rmse_2}')

y_2_hat_test = lr_2.predict(X_2_test)
rmse_2 = metrics.mean_squared_error(y_2_test, y_2_hat_test, squared= False)
print(f'Test RMSE : {rmse_2}')
````
